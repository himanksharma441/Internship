# Implementing Machine Learning through Linear Regression

```python
#Seperating target and its only required so that x only has input variables and y only has output variable
x=df.iloc[:,2:6]
y=df.iloc[:,1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, random_state=42, test_size=.20)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(x_train,y_train)
```

The above score shows that the model did good in its training phase

```python
lrpred=lr.predict(x_test)
print("Predicted Value", lrpred)
print("Actual Value", y_test)
```

The above data shows the predicted value and the actual value. We can see that there is not much difference between both the data so the model is working good in predicting Happiness Score 

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import r2_score

print('Mean Squared Error:', mse(y_test, lrpred))
print('Mean Absolute Error:', mae(y_test, lrpred))
print('Root Mean Squared Error:', np.sqrt(mse(y_test, lrpred)))
print('R2 Score:', r2_score(y_test, lrpred))
```

The model has an Root Mean Squared Error of -0.48. This implies that lower is the RMSE Score, therefore model is performing better at making predicitions
