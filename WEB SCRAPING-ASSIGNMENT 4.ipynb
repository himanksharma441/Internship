1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
viewrank=[]
viewname=[]
viewartist=[]
uploaddate=[]
videoview=[]

try:
    rank_tags=driver.find_elements(By.XPATH,'//*[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
    for r in rank_tags[0:30]:
        viewrank.append(r.text.split('.')[0])
except NoSuchElementException:
    viewrank.append('-')

try:
    vwname_tags=driver.find_elements(By.XPATH,'//*[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
    for vn in vwname_tags[0:30]:
        viewname.append(vn.text.split('[')[0])
except NoSuchElementException:
    viewname.append('-')

try:
    vart_tags=driver.find_elements(By.XPATH,'//*[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
    for va in vart_tags[0:30]:
        viewartist.append(va.text)
except NoSuchElementException:
    viewartist.append('-')
        
try:
    upload_tags=driver.find_elements(By.XPATH,'//*[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
    for up in upload_tags[0:30]:
        uploaddate.append(up.text)
except NoSuchElementException:
    uploaddate.append('-')
    
try:
    videoview_tags=driver.find_elements(By.XPATH,'//*[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
    for vd in videoview_tags[0:30]:
        videoview.append(vd.text)
except NoSuchElementException:
    videoview.append('-')
```

```python
print(len(viewrank),len(viewname),len(viewartist),len(uploaddate),len(videoview))
```

```python
df=pd.DataFrame({'Rank':viewrank,'Name':viewname,' Artist':viewartist,'Upload Date':uploaddate,'Views(billions)':videoview})
df
```

8. Details of Datasets from UCI machine learning repositories.
Url = https://archive.ics.uci.edu/
You have to find the following details:
A) Dataset name
B) Data type
C) Task
D) Attribute type
E) No of instances
F) No of attribute G) Year

```python
driver.get("https://archive.ics.uci.edu/")
```

```python
dataset=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]")
dataset.click()
```

```python
set_url=[]
url=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
for i in url:
    set_url.append(i.get_attribute("href"))
```

```python
datasetname=[]
datatype=[]
task=[]
attributetype=[]
instances=[]
noofattribute=[]
uciyear=[]

for i in set_url:
    driver.get(i)
    
    try:
        datasetname_tags=driver.find_element(By.XPATH,'//h1[@class="text-3xl font-semibold text-primary-content"]')
        datasetname.append(datasetname_tags.text)
    except NoSuchElementException:
        datasetname.append('-')

    try:
        datatype_tags=driver.find_element(By.XPATH,'//*[@class="relative flex flex-col gap-4 bg-base-100 p-4 shadow"]/div[2]/div[1]/p')
        datatype.append(datatype_tags.text)
    except NoSuchElementException:
        datatype.append('-')
    
    try:
        task_tags=driver.find_element(By.XPATH,'//*[@class="relative flex flex-col gap-4 bg-base-100 p-4 shadow"]/div[2]/div[3]/p')
        task.append(task_tags.text)
    except NoSuchElementException:
        task.append('-')
    
    try:
        attributetype_tags=driver.find_element(By.XPATH,'//*[@class="relative flex flex-col gap-4 bg-base-100 p-4 shadow"]/div[2]/div[4]/p')
        attributetype.append(attributetype_tags.text)
    except NoSuchElementException:
        attributetype.append('-')
    
    try:
        instances_tags=driver.find_element(By.XPATH,'//*[@class="relative flex flex-col gap-4 bg-base-100 p-4 shadow"]/div[2]/div[5]/p')
        instances.append(instances_tags.text)
    except NoSuchElementException:
        instances.append('-')
    
    try:
        noofattribute_tags=driver.find_element(By.XPATH,'//*[@class="relative flex flex-col gap-4 bg-base-100 p-4 shadow"]/div[2]/div[6]/p')
        noofattribute.append(noofattribute_tags.text)
    except NoSuchElementException:
        noofattribute.append('-')
    
    try:
        uciyear_tags=driver.find_element(By.XPATH,'//h2[@class="text-primary-content"]')
        uciyear.append(uciyear_tags.text)
    except NoSuchElementException:
        uciyear.append('-')
```

```python
print(len(datasetname),len(datatype),len(task),len(attributetype),len(instances),len(noofattribute),len(uciyear))
```

```python
df=pd.DataFrame({'Dataset name':datasetname,'Data type':datatype,'Task':task,'Attribute type':attributetype,'No of instances':instances,'No of attribute':noofattribute,'Year':uciyear})
df
```
